```python
import pandas as pd
t1_train=pd.read_json("/content/MaSaC_train_erc.json")
t1_val=pd.read_json("/content/MaSaC_val_erc.json")

df_val=pd.DataFrame(t1_val)
df_val=pd.DataFrame(t1_val)
```

```python
utterances_all_eps=pd.DataFrame(t1_train["utterances"])
all_utt_train=pd.DataFrame(columns=[])
for curr_ep in utterances_all_eps["utterances"]:
#     curr_ep=utterances_all_eps["utterances"][0]
    utt_of_Ith_ep=pd.DataFrame(curr_ep)
    all_utt_train=pd.concat([utt_of_Ith_ep,all_utt_train],axis=0)


utterances_all_eps=pd.DataFrame(t1_val["utterances"])
all_utt_val=pd.DataFrame(columns=[])
for curr_ep in utterances_all_eps["utterances"]:
#     curr_ep=utterances_all_eps["utterances"][0]
    utt_of_Ith_ep=pd.DataFrame(curr_ep)
    all_utt_val=pd.concat([utt_of_Ith_ep,all_utt_val],axis=0)
```

```python
emotion_all_eps=pd.DataFrame(t1_train["emotions"])
all_emo_train=pd.DataFrame(columns=[])
for curr_ep in emotion_all_eps["emotions"]:
    emo_of_Ith_ep=pd.DataFrame(curr_ep)
    all_emo_train=pd.concat([emo_of_Ith_ep,all_emo_train],axis=0)


emotion_all_eps=pd.DataFrame(t1_val["emotions"])
all_emo_val=pd.DataFrame(columns=[])
for curr_ep in emotion_all_eps["emotions"]:
    emo_of_Ith_ep=pd.DataFrame(curr_ep)
    all_emo_val=pd.concat([emo_of_Ith_ep,all_emo_val],axis=0)
```

```python
data_train_task1=pd.concat([all_utt_train,all_emo_train],axis=1)
data_val_task1=pd.concat([all_utt_val,all_emo_val],axis=1)
```

```python
data_train_task1.to_csv('data_train_task1',header=["text","label"], index=False)
```

```python
data_val_task1.to_csv('data_val_task1',header=["text","label"], index=False)
```

```python
train=pd.read_csv('/content/data_train_task1')
val=pd.read_csv('/content/data_val_task1')
```

OVERSAMPLING

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(sampling_strategy='not majority')

X_resampled, y_resampled = oversample.fit_resample(train[['text']], train['label'])
train_resampled = pd.DataFrame(X_resampled, columns=['text'])
train_resampled['label'] = y_resampled

X_resampled, y_resampled = oversample.fit_resample(val[['text']], val['label'])
val_resampled = pd.DataFrame(X_resampled, columns=['text'])
val_resampled['label'] = y_resampled
```

```python
print("After over sampling :")
print("TRAIN--->")
print(train_resampled['label'].value_counts())
print("VAl------>")
print(val_resampled['label'].value_counts())
```
