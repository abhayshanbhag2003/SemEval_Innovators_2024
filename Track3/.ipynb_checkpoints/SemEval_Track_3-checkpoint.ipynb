```python
import json

# Replace 'your_file_path.json' with the path to your JSON file
file_path = r"C:\Users\Suramya\Desktop\Shared_Task\MELD_test_efr.json"

# Using 'with' ensures the file is closed properly after opening
with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
y_pred_algo_1=[]
y_actual_algo_1=[]
y_pred_algo_2=[]
y_actual_algo_2=[]
```

```python
def trig_from_episode_bruteFroce(episodes):
  lengthofSpeakers=len(episodes['speakers'])
  trig=[]
  for i in range(lengthofSpeakers):
    if(i!=lengthofSpeakers-2):
      trig.append("0.0")
    else:
      trig.append("1.0")
  return trig,episodes['labels']
```

```python
def trig_From_Episode(Episode):
  context={}
  for i in range(len(Episode['speakers'])):
      speaker = Episode['speakers'][i]
      emotion = Episode['emotions'][i]

      if speaker not in context:
          context[speaker] = []

      context[speaker].append({emotion: i})

  for speaker, emotions_list in context.items():
      context[speaker] = emotions_list
  lastFlipForEverySpeaker=[]
  for sp in context:
    lastFlip=0
    lastEmo='null'
    for dicti in context[sp]:
      emo=list(dicti.keys())[0]
      index=dicti[emo]
      if(lastEmo!=emo):
        lastFlip=index
    # print("For",sp," flip at ", lastFlip-1)
    lastFlipForEverySpeaker.append(lastFlip-1)
  trig=[]
  for i  in range(len(Episode['speakers'])):
    if(i in lastFlipForEverySpeaker ):
      trig.append("1.0")
    else:
      trig.append("0.0")
  # # print(trig)
  # print(len(trig)," ",len(Episode['labels']))
  return trig,Episode['labels']

```

```python
for ep in data:
  trig,label=trig_from_episode_bruteFroce(ep)
  y_pred_algo_2.extend(trig)
  y_actual_algo_2.extend(label)
```

```python
for ep in data:
  trig1,label1= trig_From_Episode(ep)
  y_pred_algo_1.extend(trig1)
  y_actual_algo_1.extend(label1)
```
